```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("csvStreaming").getOrCreate()
```

```python
from pyspark.sql.types import *
schema = StructType([
    StructField("Date", DateType(), True),
    StructField("Open", DoubleType(), True),
    StructField("High", DoubleType(), True),
    StructField("Low", DoubleType(), True),
    StructField("Close", DoubleType(), True),
    StructField("Volume", IntegerType(), True),
    StructField("Adj Close", DoubleType(), True)
])

df = spark.readStream\
    .format("csv")\
    .option("header","True")\
    .schema(schema).load("file:///home/hdoop/notebooks/data/spark_practice/csvStreaming/*")
    
```

```python
query = df.writeStream.outputMode("append").format("console").start().awaitTermination()
```

```python
spark.stop()
```
